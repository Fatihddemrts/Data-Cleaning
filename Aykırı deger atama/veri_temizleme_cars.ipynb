```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.plot(x='brand', y='selling_price', style='o')
```

```python
sns.boxplot(data=df["selling_price"])
```

```python
Q1 = df['selling_price'].quantile(0.25)
Q3 = df['selling_price'].quantile(0.75)
IQR = Q3 - Q1
altsinir = Q1 - 1.5 * IQR
ustsinir = Q3 + 1.5 * IQR
```

```python
askck = df[(df['selling_price'] < altsinir) | (df['selling_price'] > ustsinir)]
askck
```

```python
df.loc[df['selling_price'] > ustsinir, 'selling_price'] = ustsinir
df.loc[df['selling_price'] < altsinir, 'selling_price'] = altsinir
```

```python
sns.boxplot(data=df["selling_price"])
plt.show()
```
