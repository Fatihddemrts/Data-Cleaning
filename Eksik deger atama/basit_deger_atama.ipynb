```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.nan,7,1,np.nan,9,15])
V2 = np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
V3 = np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df = pd.DataFrame(
    {
        "V1": V1,
        "V2": V2,
        "V3": V3
    }
)
df
```

Sayısal degiskenlerde atama


```python
df["V1"].fillna(0)##nan olanları 0 ile doldurur
```

```python
df
```

```python
df["V1"].fillna(df["V1"].mean())##nan olanları ortalama ile doldurur
```

```python
## Tum degiskenler icin birinci yol
df.apply(lambda x: x.fillna(x.mean()), axis=0)##her bir değişkenin ortalaması ile doldurur
```

```python
## ikinci yol   
```

```python
df.fillna(df.mean()[:])##her bir değişkenin ortalaması ile doldurur
```

```python
df
```

```python
df.fillna(df.mean()["V1":"V2"])##V1 ve V2 değişkenlerinin ortalaması ile doldurur
```

```python
df["V3"].fillna(df["V3"].median())##nan olanları medyan ile doldurur
```

```python
## ucuncu yol
```

```python
df.where(pd.notna(df), df.mean(), axis="columns")##her bir değişkenin ortalaması ile doldurur
```
