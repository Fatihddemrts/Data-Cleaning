```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.nan,7,1,np.nan,9,15])
V2 = np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
V3 = np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df = pd.DataFrame(
    {
        "V1": V1,
        "V2": V2,
        "V3": V3
    }
)
df
```

```python
df.dropna()##tüm satırların hepsinin eksik değer içermesi durumunda silme işlemi yapar
```

```python
df
```

```python
df.dropna(how = "all")##tüm satırların hepsinin eksik değer içermesi durumunda silme işlemi yapar
```

```python
df.dropna(axis=1)##tüm sütunların hepsinin eksik değer içermesi durumunda silme işlemi yapar
```

```python
df.dropna(axis=1, how = "all")##tüm sütunların hepsinin eksik değer içermesi durumunda silme işlemi yapar
```
